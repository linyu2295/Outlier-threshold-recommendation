```python
import json
import numpy as np
import pandas as pd
import glob
import os
import time
import datetime
# import plotly.express as px

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
filename_list = glob.glob(os.path.join('/Users/lin/Desktop/VZ_projects/Outlier/', 'part*'))
```

```python
start_time = time.time()

dat = {}
for filename in filename_list:
    tmp_time = time.time()
    dat[filename] = []
    with open(filename) as f:
        for line in f:
            dat[filename].append(json.loads(line))
    print('finished loading {} file in {}seconds'.format(filename, round(time.time()-tmp_time, 2)))

print('running time: {}seconds'.format(time.time()-start_time))
```

```python
dat.keys()
```

```python
df1 = pd.DataFrame(columns=['date', 'property_id', 'CPV'])

for filename in sorted(filename_list)[:4]:
    print(filename)
    tmp_df = pd.DataFrame(dat[filename])
    tmp_df['CPV'] = tmp_df['metrics'].map(lambda x: x[0]['value'] if x[0]['name'] == '1202' else np.nan)
    tmp_df.drop(['metrics'], axis = 1, inplace = True)
    
    df1 = pd.concat([df1, tmp_df], axis = 0).reset_index(drop=True)
    print(tmp_df.shape, df1.shape)

    
df2 = pd.DataFrame(columns=['date', 'property_id', 'CPV'])

for filename in sorted(filename_list)[4:8]:
    print(filename)
    tmp_df = pd.DataFrame(dat[filename])
    tmp_df['CPV'] = tmp_df['metrics'].map(lambda x: x[0]['value'] if x[0]['name'] == '1202' else np.nan)
    tmp_df.drop(['metrics'], axis = 1, inplace = True)
    
    df2 = pd.concat([df2, tmp_df], axis = 0).reset_index(drop=True)
    print(tmp_df.shape, df2.shape)
    
    
df3 = pd.DataFrame(columns=['date', 'property_id', 'CPV'])

for filename in sorted(filename_list)[8:]:
    print(filename)
    tmp_df = pd.DataFrame(dat[filename])
    tmp_df['CPV'] = tmp_df['metrics'].map(lambda x: x[0]['value'] if x[0]['name'] == '1202' else np.nan)
    tmp_df.drop(['metrics'], axis = 1, inplace = True)
    
    df3 = pd.concat([df3, tmp_df], axis = 0).reset_index(drop=True)
    print(tmp_df.shape, df3.shape)
        
```

```python
dates = sorted(['2020-09-29', '2020-09-25', '2020-09-24', '2020-09-28',
       '2020-09-26', '2020-09-27', '2020-09-23'])
dates
```

```python
'''
    df_date: dictionary
        - key: dates
        - value: dataframe consists of date, property_id, CPV
    
'''
# df1
df_date = {}
for d in dates:
    df_date[d] = df1[df1['date'] == d].reset_index(drop=True)

# df2
for d in dates:
    print(df_date[d].shape)
    tmp_date = df2[df2['date'] == d].reset_index(drop=True)
    print(tmp_date.shape)
    
    tmp_date = pd.concat([df_date[d], tmp_date]).reset_index(drop=True)
    df_date[d] = tmp_date
    print(df_date[d].shape)
    
# df3
for d in dates:
    print(d)
    print(df_date[d].shape)
    tmp_date = df3[df3['date'] == d].reset_index(drop=True)
    print(tmp_date.shape)
    
    tmp_date = pd.concat([df_date[d], tmp_date]).reset_index(drop=True)
    df_date[d] = tmp_date
    print(df_date[d].shape, '\n')
```

### Read CPV outlier thresholds

```python
'''
    table_cpv_thresholds:
        - current CPV thresholds table
        
'''
table_cpv_thresholds = pd.read_csv('CPV outlier thresholds - raw.csv')
print(table_cpv_thresholds.shape)

table_cpv_thresholds.head()
```

```python
'''
    New
    
    
    res_propertyid_dow_with_new_threshold: 
        - property_id
        - date_record
        - day_in_week
        - CPV cnts
        - raw_threshold
        - percentile_threshold(%)

        - new threshold (0.1%)
        - new threshold (0.05%)
        - new threshold (0.01%)



'''
res_propertyid_dow_with_new_threshold = pd.DataFrame(columns=['property_id', 'date_record', 
                                           'day_in_week', 'CPV cnts', 
                                           'raw_threshold', 'percentile_threshold(%)',
                                           'new_threshold(0.1%)', 'num_CPVs_removed(0.1)%', 'num_users_removed(0.1)%',
                                           'new_threshold(0.05%)', 'num_CPVs_removed(0.05)%', 'num_users_removed(0.05)%',
                                           'new_threshold(0.01%)', 'num_CPVs_removed(0.01)%', 'num_users_removed(0.01)%'])

property_id_lists = sorted(table_cpv_thresholds['property_id'].unique())
print(len(property_id_lists))

start_time = time.time()

for d in dates:
#     d = dates[0]
    print(d)
    # p_id = property_id_lists[0]

    for p_id in property_id_lists:
        tmp_time = time.time()

        # get day of week 
        year, month, day = (int(x) for x in d.split('-'))   
        dow = datetime.date(year, month, day).strftime("%A")
        print('Day of week: {} and property id: {}'.format(dow, p_id))

        tmp_df = df_date[d][df_date[d]['property_id'] == str(p_id)]
        print('Dimension of subset for property id {}: {}'.format(p_id, tmp_df.shape))

        try:
            tmp_threshold_val = table_cpv_thresholds[(table_cpv_thresholds['property_id'] == p_id)&\
                                                    (table_cpv_thresholds['day_in_week'] == dow[:3])]['threshold'].values[0]

            tmp_percentile = (sum(tmp_df['CPV'] <= tmp_threshold_val)/tmp_df.shape[0])*100
            print('Threshold value: {} and {}th percentile'.format(tmp_threshold_val, round(tmp_percentile, 2)))

            res_propertyid_dow_with_new_threshold = res_propertyid_dow_with_new_threshold.append(pd.Series([str(p_id), d, dow[:3], 
                                                                      tmp_df.shape[0], tmp_threshold_val, tmp_percentile,
                                                                      round(tmp_df['CPV'].quantile(0.999),0), 
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.999),0)]['CPV'].sum(),
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.999),0)].shape[0],
                                                                      round(tmp_df['CPV'].quantile(0.9995),0),
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.9995),0)]['CPV'].sum(),
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.9995),0)].shape[0],
                                                                      round(tmp_df['CPV'].quantile(0.9999),0),
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.9999),0)]['CPV'].sum(),
                                                                        tmp_df[tmp_df['CPV'] > round(tmp_df['CPV'].quantile(0.9999),0)].shape[0]], 
                                                index=res_propertyid_dow_with_new_threshold.columns), ignore_index=True)
        except:
            print('No property id {} found!'.format(p_id))

        print('Finished percentile calculation for property id {} with {}seconds'.format(p_id, 
                                                                        round(time.time()-tmp_time, 2)), '\n')
    print('\n')
        
print('Total running time: {}seconds'.format(round(time.time()-start_time, 2)))
```

```python
res_propertyid_aggregated = pd.DataFrame(columns=['property_id', 
                                                  '7days user cnts', 
                                                  'raw_threshold_mean',
                                                  'raw_threshold_median',
                                                  'mean_percentile_threshold(%)',
                                                  'median_percentile_threshold(%)',
                                                  'new_threshold(0.3%)', 'num_CPVs_removed(0.3)%', 'num_users_removed(0.3)%',
                                                  'new_threshold(0.25%)', 'num_CPVs_removed(0.25)%', 'num_users_removed(0.25)%',
                                                  'new_threshold(0.2%)', 'num_CPVs_removed(0.2)%', 'num_users_removed(0.2)%',
                                                  'new_threshold(0.15%)', 'num_CPVs_removed(0.15)%', 'num_users_removed(0.15)%',
                                                  'new_threshold(0.1%)', 'num_CPVs_removed(0.1)%', 'num_users_removed(0.1)%',
                                                  'new_threshold(0.05%)', 'num_CPVs_removed(0.05)%', 'num_users_removed(0.05)%',
                                                  'new_threshold(0.01%)', 'num_CPVs_removed(0.01)%', 'num_users_removed(0.01)%'])
```

```python
property_id_lists = sorted(table_cpv_thresholds['property_id'].unique())
print(len(property_id_lists))

# p_id = property_id_lists[0]
# print(p_id)

for p_id in property_id_lists:

#     if p_id in table_cpv_thresholds['property_id']:
    all_df = pd.DataFrame(columns=['CPV', 'date', 'property_id'])
    tmp_time = time.time()

    for d in dates:

        # get day of week 
        year, month, day = (int(x) for x in d.split('-'))   
        dow = datetime.date(year, month, day).strftime("%A")
        print('Day of week: {} and property id: {}'.format(dow, p_id))

        if str(p_id) in df_date[d]['property_id'].unique():
            tmp_df = df_date[d][df_date[d]['property_id'] == str(p_id)]
            print('Dimension of subset for property id {}: {}'.format(p_id, tmp_df.shape))

            all_df = pd.concat([all_df, tmp_df], ignore_index=True)
#             print(all_df.shape, '\n')
    
    if all_df.shape[0] > 0:
        print(all_df.shape)
    
        tmp_threshold_mean = table_cpv_thresholds[table_cpv_thresholds['property_id'] == p_id]['threshold'].mean()
        tmp_threshold_median = table_cpv_thresholds[table_cpv_thresholds['property_id'] == p_id]['threshold'].quantile(0.5)

        tmp_mean_percentile = (sum(all_df['CPV'] > tmp_threshold_mean)/all_df.shape[0])*100
        tmp_median_percentile = (sum(all_df['CPV'] > tmp_threshold_median)/all_df.shape[0])*100

        print('mean and median:', tmp_threshold_mean, tmp_threshold_median)
        print('percentiles: ', tmp_mean_percentile, tmp_median_percentile)

        res_propertyid_aggregated = res_propertyid_aggregated.append(pd.Series([str(p_id),
                                                                              all_df.shape[0],
                                                                              tmp_threshold_mean,
                                                                              tmp_threshold_median,
                                                                              tmp_mean_percentile,
                                                                              tmp_median_percentile,
                                                                              round(all_df['CPV'].quantile(0.997), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.997), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.997), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.9975), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9975), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9975), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.998), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.998), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.998), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.9985), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9985), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9985), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.999), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.999), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.999), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.9995), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9995), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9995), 0)].shape[0],
                                                                              round(all_df['CPV'].quantile(0.9999), 0),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9999), 0)]['CPV'].sum(),
                                                                              all_df[all_df['CPV']>round(all_df['CPV'].quantile(0.9999), 0)].shape[0]
                                                                              ],
                                                                             index=res_propertyid_aggregated.columns), 
                                                                    ignore_index=True)

        print('Finished percentile calculation for property id {} with {}seconds'.format(p_id, 
                                                                    round(time.time()-tmp_time, 2)), '\n')

    else:
        print('No property id {} found!'.format(p_id), '\n')
```
